```python
import pandas as pd
from pathlib import Path
from sklearn.model_selection import train_test_split
```

```python
IN = 'data/features_3_sec.csv'
OUT_TRAIN = 'data/train_3_sec.csv'
OUT_TEST = 'data/test_3_sec.csv'
```

```python
df = pd.read_csv(IN)
```

```python
df[(df.isna().any(axis=1)) | ((df == 0).any(axis=1))]
```

```python
df_tr, df_te = train_test_split(df, test_size=0.2, random_state=42, shuffle=True)
```

```python
Path(OUT_TRAIN).parent.mkdir(parents=True, exist_ok=True)
df_tr.to_csv(OUT_TRAIN, index=False)
print(f"Done! -> {OUT_TRAIN}  rows={len(df_tr)}  cols={df_tr.shape[1]}")
```

```python
Path(OUT_TEST).parent.mkdir(parents=True, exist_ok=True)
df_te.to_csv(OUT_TEST, index=False)
print(f"Done! -> {OUT_TEST}  rows={len(df_te)}  cols={df_te.shape[1]}")
```
